```python
data['AppName'].unique()
dict={'OnePlan':1, 'VPlan':2, 'ICGS':3, 'COEP':4, 'GFIM':5, 'PRS':6, 'WFM':7, 'COEA':8,
       'vATOM':9, 'DNM':10}
data['mAppName']=data['AppName'].map(dict)
dict1={"Pass":1,"Fail":0}
data['mSTATUS']=data['STATUS'].map(dict1)

data['mEXEDATE']=data['EXEDATE'].apply(lambda x: str(x).replace('.',""))
data['mEXEDATE']
```

```python
data['ENV'].unique()
```

```python
dict3={'PROD2.0': 101, 'DEV2.0': 102, 'STG2.0': 103, 'UAT2.0': 104}
data['mENV']=data['ENV'].map(dict3)
data
```

```python
feature_cols = ['mAppName', 'TCID','mENV']
X = data[feature_cols] 
y = data.mSTATUS
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=1)
print(X_train)
```

```python
clf =LogisticRegression(  class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=110, multi_class='ovr', n_jobs=1,
          penalty='l1', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
clf = clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
len(y_pred)
len(y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
